# Mean, Median and Mode

```python
# Create an income dataset with 10000 datapoints/records 
# centered aroud 27000 and
# standard deviation 15000
import numpy as np
income = np.random.normal(27000, 15000, 10000)
income
```

```python
for i in range(0,10000):
    print(income[i])
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.hist(income, 50)
plt.show()
```

```python
np.mean(income)
```

```python
np.median(income)
```

```python
 income = np.append(income, [1000000000])
```

```python
np.mean(income)
```

```python
np.median(income)
```

### Mode
