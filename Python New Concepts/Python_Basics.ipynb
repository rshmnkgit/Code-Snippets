# Bitwise Operators

Bitwise operators
Bitwise operators act on operands as if they were strings of binary digits. They operate bit by bit, hence the name.

For example, 2 is 10 in binary and 7 is 111.

In the table below: Let x = 10 (0000 1010 in binary) and y = 4 (0000 0100 in binary)




```python
bin(5),  bin(6),  bin(4),  bin(7),  bin(3)
```

```python
# AND
5 & 6                   # 101  &  110  =   100
```

```python
# OR
5 | 6                   # 101  |  110   =  111
```

```python
# XOR          Sets each bit to 1 if only one of two bits is 1

5 ^ 6          # 101 ^ 110 = 011
```

```python
# NOT

~10        # ~101 = 010
```

```python
def set_alarm(employed, vacation):
    # Your code here
    if employed:
        if vacation:
            return False
        else:
            return True
    else:
        return False

set_alarm(1,1)
```

```python
# setAlarm(true, true) -> false
# setAlarm(false, true) -> false
# setAlarm(false, false) -> false
# setAlarm(true, false) -> true
```

```python
s = "abcd"
position = 0
result = ''
for letter in s:
    lu = letter.upper()
    ls = letter.lower() * position
    if (result == ''):
        result = result + lu + ls
    else:
        result = result + "-" + lu + ls
    print(result)
    position += 1
```

```python
r=""
for i,c in enumerate(s):
    if (r != ''):
        r = r + "-"
    r = r + c.upper() + c.lower() * i 
    print(i, c, r)
```
