```python
def almostIncreasingSequence(sequence):
    droppped = False
    last = prev = min(sequence) - 1
    for elm in sequence:
        if elm <= last:
            if droppped:
                return False
            else:
                droppped = True
            if elm <= prev:
                prev = last
            elif elm >= prev:
                prev = last = elm
        else:
            prev, last = last, elm
    return True
```

```python
# Return the maximum of the products of adjacent elements in an array

def adjacentElementsProduct(inputArray):
    maxProd = 0
    for i in len(inputArray):
        prod = inputArray[i] * inputArray[i+1]
        if prod > maxProd:
            maxProd = prod
    return maxProd
```

```python
def shopInAndOutEvents(events):
    # get the number of events
    eset = set()
    for name, status in events:
        if name in eset and status =="out":
            eset.remove(name)
        elif name in eset and status =="in":
            return False
        elif name not in eset and status=="out":
            return False
        elif name not in eset and status=="in":
            eset.add(name)

    if len(eset)>0:
        return False
            
    return True
            

pevents = [["John", "in"],["John", "out"],["Mary","in"],["Mary","out"],["John", "in"],["John", "out"]]


print(shopInAndOutEvents(pevents))
```
