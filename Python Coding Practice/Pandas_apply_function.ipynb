```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("resources/bigmart_data.csv")
df.head()
```

# apply() 

```python
# Access all rows, one by one row wise
df.apply(lambda x : x)
```

```python
# Access a single row
df.apply(lambda x : x[4])
```

```python
# Access a single column
df.apply(lambda x : x[0], axis = 1)
```

```python
# Access using column name
df.apply(lambda x : x['Item_Fat_Content'], axis=1)
```

```python
def clip_price(price):
    if price>200:
        price = 200
    return price
```

```python
df.head()
```

```python
df['Item_MRP'].apply(lambda x : clip_price(x))
```

```python
df
```

```python
df['Item_MRP'].apply(lambda x : 200 if x>200 else x)
```

```python
def label_encode(city):
    if city == 'Tier 1':
            label = 0
    elif city == 'Tier 2':
            label = 1
    else:
            label = 2
    return label
    
```

```python
df['Outlet_Location_Type'].apply(lambda x : label_encode(x))
```

# Aggregating Data

#### What is the mean price of each item type?

```python
price_by_item =df.groupby('Item_Type')
# price_by_item
price_by_item.first()
```

```python
price_by_item['Item_MRP'].mean()
```

```python
price_by_item[['Item_MRP', 'Item_Fat_Content']]
price_by_item.first()
```

# Crosstab

```python
pd.crosstab(df['Outlet_Size'], df['Outlet_Location_Type'])
```

```python
pd.crosstab(df['Outlet_Size'], df['Outlet_Location_Type'], margins=True)
```

# Pivot Table

```python
pd.pivot_table(df, index=['Outlet_Establishment_Year'], values=['Item_Outlet_Sales'])
```

```python
pd.pivot_table(df, index=['Outlet_Establishment_Year', 'Outlet_Location_Type', 'Outlet_Size'], values=['Item_Outlet_Sales'])
```

```python
pd.pivot_table(df, index=['Outlet_Establishment_Year', 'Outlet_Location_Type', 'Outlet_Size'], values=['Item_Outlet_Sales'],\
              aggfunc=[np.mean, np.median, min, max, np.std])
```
